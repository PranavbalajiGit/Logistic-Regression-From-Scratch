```python
import numpy as np
```

```python
class Logistic_Regression():
    def __init__(self, learning_rate , no_of_iterations):
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations        

    def fit(self , X , Y):
        
        self.m , self.n = X.shape

        self.X = X
        self.Y = Y

        self.w = np.zeros(self.n)
        self.b = 0

        for i in range(self.no_of_iterations):
            self.update_weights()

    def update_weights(self):
        
        #Here we didn't use predict funtion because we need to need train the model
        #with Exact values. While using test data it need to result either 1 or 0
        #So at that time we use predict() function to get the Output. 
        Y_hat = (1 / (1 + np.exp(-(self.X.T.dot(self.w) + self.b))))

        #This Partial Derivative is used in Logistic Regression for 
        #Binary Cross Entropy Loss Function
        dw = (np.dot(self.X.T , (Y_hat - self.Y))) / self.m
        db = (np.sum(Y_hat - self.Y)) / self.m

        self.w = self.w - self.learning_rate * dw
        self.b = self.b - self.learning_rate * db


    def predict(self , X):
        #Here it is not self.X it is just X because we use this functions argument X data. 
        Y_pred = ((1 / (1 + np.exp(-(X.T.dot(self.w) + self.b)))))
        Y_pred = np.where(Y_pred > 0.5 , 1 , 0)
        return Y_pred
        
```
